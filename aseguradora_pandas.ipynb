# estadistica descriptiva

```python
resumen_estadistico_imc = df['imc'].describe()
print(resumen_estadistico_imc)
```

```python
#frecuencia absoluta
frecuencia_region = df['region'].value_counts()
print(frecuencia_region)
```

```python
#moda
moda_imc  = df['imc'].mode()
print(moda_imc)
```

```python
import matplotlib.pyplot as plt

df['region'].value_counts().plot(kind='bar', color=['red', 'green', 'blue', 'yellow'])
plt.title("frecuencia  DE REGIONES")
plt.xlabel("REGIONES")
plt.ylabel("FRECUENCIA")
plt.show()
```
